```python
from nltk.parse import CoreNLPParser
from nltk.corpus.reader.conll import ConllCorpusReader
from nltk.tag.perceptron import PerceptronTagger
from nltk.grammar import DependencyGrammar
from nltk.parse import (
     DependencyGraph,
     ProjectiveDependencyParser,
     NonprojectiveDependencyParser,
 )
from nltk import FreqDist
from pprint import pprint
import math

parser = CoreNLPParser(url='http://localhost:9000')

root = "."
train = "fr-ud-train.conllu3"
test = "fr-ud-test.conllu3"
COLUMN_TYPES = ('ignore', 
                'words', 
                'ignore', 
                'pos', 
                'ignore', 
                'ignore', 
                'ignore', 
                'ignore', 
                'ignore', 
                'ignore')

trainCorpus  = ConllCorpusReader(root=root, 
                                  fileids=train, 
                                  columntypes=COLUMN_TYPES, 
                                  encoding='utf8', 
                                  separator="\t", 
                                  tagset='universal')

testCorpus  = ConllCorpusReader(root=root, 
                                  fileids=test, 
                                  columntypes=COLUMN_TYPES, 
                                
                                  encoding='utf8', 
                                  separator="\t", 
                                  tagset='universal')

testWords = testCorpus.tagged_sents()

myfile = open("./fr-ud-test.conllu3")
conllfile = myfile.read()

# séparation du fichier conll en lignes
ff = conllfile.split('\n\n')

dg = []
sents = []

# génération des graphes à partir des lignes du fichier conll
for f in ff[1:4]:
    words = []
    print(f)
    o = f.split('\n')
    

    for j in range(0,len(o)):
        sen = o[j].split('\t')
        words.append(sen[1] + ' ')
        print(words)
    sents.append(''.join(words))
print(sents)

nps = []

for sent in sents[:1]:
    tree = next(parser.raw_parse(sent))
    for st in tree.subtrees(filter = lambda x: x.label() == "NP"):
        nps.append(st)

fdist = FreqDist(nps)
fdist.most_common(10)

```
